```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import pandas as pd
```

## 1. Go to your favorite actress'/actor's imdb page

```python
#THE 211 BEST NETFLIX SERIES
base_site = "https://editorial.rottentomatoes.com/guide/best-netflix-shows-and-movies-to-binge-watch-now/"

response = requests.get(base_site)
response.status_code
```

```python
html = response.content
soup = BeautifulSoup(html, "lxml")
```

## 2. Get all the links to his person's movies

```python
raw = soup.find_all("div",{"class":"article_movie_title"})
```

```python
len(raw)
```

```python
links = [i.find("a").get("href") for i in raw]
```

```python
len(links)
```

```python
links_ordered = ["https:"+ i for i in links]
links_ordered
```

## 3. Get each movie's name, director and imdb score

### Names

```python
names = [i.find("a").text for i in raw]
```

```python
names_df = pd.DataFrame(names, columns = ["Names"])
names_df
```

### Years

```python
years = [i.find("span",{"class":"subtle start-year"} ).text.strip("()") for i in raw]
```

```python
years_df = pd.DataFrame(years, columns = ["Years"] )
years_df
```

### Scores

```python
scores = ["%"+i.find("span",{"class","tMeterScore"} ).text[:-1] for i in raw]
```

```python
scores_df = pd.DataFrame(scores, columns = ["Scores"])
scores_df
```

```python
xx = pd.concat([names_df, years_df, scores_df], axis = 1)
xx
```

```python
xx = xx[["Scores", "Names", "Years"]]
```
