## 2. User-Based Recommender Systems

### Data Preparation

```python
df_agg
```

```python
product_sold_counts = pd.DataFrame(df_agg["Code_Product"].value_counts())
product_sold_counts
```

```python
rare_products = product_sold_counts[product_sold_counts["Code_Product"] <= 100]
rare_products
```

```python
def create_user_product_df():
    product_counts = pd.DataFrame(df_agg["Code_Product"].value_counts())
    rare_products = product_sold_counts[product_counts["Code_Product"] <= 100].index
    sold_products = df_agg[~df_agg["Code_Product"].isin(rare_products)]
    user_product_df = sold_products.pivot_table(index=["Main_ID"], columns=["Code_Product"], values="Amount", aggfunc=np.sum)
    return user_product_df

user_product_df = create_user_product_df()
random_user = (pd.Series(user_product_df.index).sample(1, random_state=45).values)[0]
# random_user = "c983862a"
random_user
```

### Identification of Users with the Most Similar Behaviors to the User to be Suggested --- Ã–neri YapÄ±lacak KullanÄ±cÄ± ile En Benzer DavranÄ±ÅŸlÄ± KullanÄ±cÄ±larÄ±n Belirlenmesi

```python
# 1. Combining data of random user and other users --- Rastgele kullanÄ±cÄ±nÄ±n(random user) ve diÄŸer kullanÄ±cÄ±larÄ±n verilerini bir araya getirme
```

```python
random_user_df[products_sold]
```

```python
products_sold_df[products_sold_df.index.isin(users_same_products)]
```

```python
final_df = pd.concat([products_sold_df[products_sold_df.index.isin(users_same_products)],
                      random_user_df[products_sold]])
final_df
```

```python
# 2.Creating the correlation df --- Korelasyon df'ini oluÅŸturma
```

```python
corr_df = final_df.T.corr().unstack().sort_values().drop_duplicates()
corr_df = pd.DataFrame(corr_df, columns=["Corr"])

corr_df.index.names = ['user_id_1', 'user_id_2']

corr_df = corr_df.reset_index()
corr_df
```

```python
# 3. Finding the most similar users (Top Users) --- En benzer kullanÄ±cÄ±larÄ± (Top Users) bulma
```

```python
top_users = corr_df[(corr_df["user_id_1"] == random_user) & (corr_df["Corr"] >= 0.65)][["user_id_2", "Corr"]].reset_index(drop=True)
top_users = top_users.sort_values(by='Corr', ascending=False)
top_users.rename(columns={"user_id_2": "Main_ID"}, inplace=True)
top_users
```

```python
df_agg
```

```python
top_users_ratings = top_users.merge(df_agg[["Main_ID", "Code_Product", "number_of_amounts"]], how='inner')

top_users_ratings = top_users_ratings[top_users_ratings["Main_ID"] != random_user]

top_users_ratings
```

### Calculation of the Weighted Average Recommendation Score ---- Weighted Average Recommendation Score'un HesaplanmasÄ±

```python
top_users_ratings['weighted_number_of_amounts'] = top_users_ratings['Corr'] * top_users_ratings['number_of_amounts']

top_users_ratings
```

```python
top_users_ratings.groupby('Code_Product').agg({"weighted_number_of_amounts": "mean"})

top_users_ratings
```

```python
recommendation_df = top_users_ratings.groupby('Code_Product').agg({"weighted_number_of_amounts": "mean"})
recommendation_df = recommendation_df.reset_index()
products_to_be_recommend = recommendation_df[recommendation_df["weighted_number_of_amounts"] > 2000].sort_values("weighted_number_of_amounts", ascending=False)

print(f"************* RECOMMENDED PRODUCTS for user with '{random_user}' user_id are:")
products_to_be_recommend
```
