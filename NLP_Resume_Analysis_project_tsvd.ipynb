```python
# !pip install nltk
import nltk     # naturel language toolkit
from nltk.tokenize import word_tokenize
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.stem import PorterStemmer
nltk.download("punkt")
nltk.download("stopwords")
from nltk.corpus import stopwords
nltk.download('wordnet')
from collections import Counter

stop_words = stopwords.words('english')
```

```python
def cleaning_tokanize(data):
       
    #1. Tokenize
    text_tokens = word_tokenize(data.lower())
                        
    #joining
    return " ".join(text_tokens)
```

```python
def cleaning_lemmatize(data):
       
    #1. Tokenize
    text_tokens = word_tokenize(data.lower())
        
    #2. Remove Puncs
    tokens_without_punc = [w for w in text_tokens if w.isalpha()]  
        # punctuation ve no lardan kurtuluyoruz.
    
    #3. Removing Stopwords
    tokens_without_sw = [t for t in tokens_without_punc if t not in stop_words]  
       
    #4.2 lemma
    text_cleaned = [WordNetLemmatizer().lemmatize(t) for t in tokens_without_sw]
                
    #joining
    return " ".join(text_cleaned)
```

## Dimension Reduction -- Truncated SVD

```python
print("Original X_train   : ", df.job_description.shape)
print("Reshaped X_train_tf: ", X_tf.shape)
```

```python
# 5 of the number of columns corresponds to a Sum of explained variance ratio of 83%.
# 3.5 of the number of columns corresponds to a Sum of explained variance ratio of over 95%.

nn = int(X_tf.shape[1]/3.5)
nn
```

```python
%%time

from sklearn.decomposition import TruncatedSVD

# Create and run an TSVD with one less than number of features
tsvd = TruncatedSVD(n_components=nn, random_state=42)
X_tf_tsvd = tsvd.fit_transform(X_tf)

# List of explained variances
tsvd_var_ratios = tsvd.explained_variance_ratio_
print("------------------------------------------------------------") 
print(f"Sum of explained variance ratio: {round(tsvd.explained_variance_ratio_.sum(),4)}")


def select_n_components(var_ratio, goal_var: float) -> int:
    
    total_variance = 0.0    # Set initial variance explained so far
    
    n_components = 0        # Set initial number of features
    
    # For the explained variance of each feature:
    for explained_variance in var_ratio:
        
        # Add the explained variance to the total
        total_variance += explained_variance
        
        # Add one to the number of components
        n_components += 1
        
        # If we reach our goal level of explained variance
        if total_variance >= goal_var:
            # End the loop
            break
            
    # Return the number of components
    return n_components

print("select_n_components: ", select_n_components(tsvd_var_ratios, 0.95))
selected_n_component = select_n_components(tsvd_var_ratios, 0.95)
```

```python
%%time

from sklearn.decomposition import LatentDirichletAllocation

# Use LDA to look for 4 topics
n_topics = 4
model_lda = LatentDirichletAllocation(n_components=n_topics,random_state=42)
model_lda.fit(X_tf_tsvd_sc)

# Print the top 4 words per topic
n_words = 4
feature_names = tf_idf.get_feature_names_out()

topic_list = []
for topic_idx, topic in enumerate(model_lda.components_):
    top_n = [feature_names[i]
              for i in topic.argsort()
               [-n_words:]][::-1]
    
    top_features = ' '.join(top_n)
    
    topic_list.append(f"topic_{'_'.join(top_n[:3])}") 

    print(f"Topic {topic_idx}: {top_features}")
    print('\n\n\n')
    
topic_result = model_lda.transform(X_tf_tsvd_sc)
df['topic']  = topic_result.argmax(axis=1)
```

```python
%%time

from sklearn.decomposition import LatentDirichletAllocation
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

# Define Search Param
# search_params = {'n_components': [10, 15, 20, 25, 30], 'learning_decay': [.5, .7, .9]}
search_params = {'n_components': [2, 3, 4, 5], 'learning_decay': [.5, .7, .9]}


# Initialize the Model
lda = LatentDirichletAllocation()

# Initiliaze Grid Search Class
model = RandomizedSearchCV(lda, param_distributions=search_params, n_jobs=-1)

# Do the Grid Search
model.fit(X_tf_tsvd_sc)
print("n_topics: ", list(model.best_params_.values())[0])
print("Best LDA model's params" , model.best_params_)
print("Best log likelihood Score for the LDA model",model.best_score_)
# print("LDA model Perplexity on train data", model.perplexity(X_cv)) # just for GridSearchCV
```

```python
def format_topics_sentences(ldamodel=None, corpus=corpus, texts=data):
    # Init output
    sent_topics_df = pd.DataFrame()

    # Get main topic in each document
    for i, row_list in enumerate(ldamodel[corpus]):
        row = row_list[0] if ldamodel.per_word_topics else row_list            
        # print(row)
        row = sorted(row, key=lambda x: (x[1]), reverse=True)
        # Get the Dominant topic, Perc Contribution and Keywords for each document
        for j, (topic_num, prop_topic) in enumerate(row):
            if j == 0:  # => dominant topic
                wp = ldamodel.show_topic(topic_num)
                topic_keywords = ", ".join([word for word, prop in wp])
                sent_topics_df = sent_topics_df.append(pd.Series([int(topic_num), round(prop_topic,4), topic_keywords]), ignore_index=True)
            else:
                break
    sent_topics_df.columns = ['Dominant_Topic', 'Perc_Contribution', 'Topic_Keywords']

    # Add original text to the end of the output
    contents = pd.Series(texts)
    sent_topics_df = pd.concat([sent_topics_df, contents], axis=1)
    return(sent_topics_df)


df_topic_sents_keywords = format_topics_sentences(ldamodel=lda_model, corpus=corpus, texts=dataa)

# Format
df_dominant_topic = df_topic_sents_keywords.reset_index()
df_dominant_topic.columns = ['Document_No', 'Dominant_Topic', 'Topic_Perc_Contrib', 'Keywords', 'Text']
df_dominant_topic.head(5)
```
